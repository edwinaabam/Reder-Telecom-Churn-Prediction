### Encoding variables

```python
'''
data.select_dtypes(include=['object']).dtypes
data.select_dtypes(include= 'datetime').dtypes
data.select_dtypes(include= 'int64').dtypes
data.select_dtypes(include= 'category').dtypes
'''
```

```python
nominal_cols= ['Gender', 'Frequency', 
               'FirstActionType', 'TotalInteractionType', 'PaymentTypes',
               'FirstInteractionType', 'LastInteractionType'] # ohe

high_cardinal_cols = ['Location',  'ProductList', 'MostCommonAction', 
                'LastPageVisited', 'FirstPageVisited', 'LastActionType', 'LeastFrequentAction'] #mean encoding


ordinal_cols = ['Segment', 'Plan','customer_segments'] #label encoding 

# target encode the ChurnLabel : 'binary'

```

```python
data[['Segment']].value_counts()
```

#### one hot encoding: nominal columns

```python
# encode nominal cols
oh_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
oh_encoded = oh_encoder.fit_transform(data[nominal_cols])

ohe_df = pd.DataFrame(oh_encoded, columns=oh_encoder.get_feature_names_out(nominal_cols))

ohe_df
```

#### TargetEncoding categorical, high cardinal columns

```python
# Target encode high cardinality columns
target_encoder = TargetEncoder(categories='auto', target_type='continuous', smooth='auto', random_state=42)

target_encoder.fit(data[high_cardinal_cols], data['TotalPurchaseValue'])
target_encoded = target_encoder.transform(data[high_cardinal_cols])

target_df = pd.DataFrame(target_encoded, columns=target_encoder.get_feature_names_out(input_features=high_cardinal_cols))

target_df
```

#### label encoding ordinal columns

```python
# Initialize and fit LabelEncoder
le = LabelEncoder()

le_df=pd.DataFrame(columns= ordinal_cols)

for c in ordinal_cols:
    le_df[c] = le.fit_transform(data[c])

le_df
```

### drop redundant columns 

```python
#drop irrelevant columns and original nominal columns that were one hot encoded: ohe_df created new differnet col names
data.drop(columns=['CustomerID', 'Name', 'Age', 'Email', 'Phone', 'Address', 
                   'Gender', 'Frequency', 
               'FirstActionType', 'TotalInteractionType', 'FirstInteractionType', 'LastInteractionType',
                 'PaymentTypes', 'Comment', 'PaymentTypes',
               'Location',  'ProductList', 'MostCommonAction', 
                'LastPageVisited', 'FirstPageVisited', 'LastActionType', 
                'LeastFrequentAction','Segment', 'Plan','customer_segments'
                ], axis=1, inplace = True)
```

```python
data.columns
```

```python
data.info()
```

### concatenate original data and all encoded columns

```python
encoded_data = pd.concat([ohe_df, data, target_df, le_df], axis=1)
```

```python
encoded_data.info()
```

### Engineer date columns

```python
# Engineer Date columns

date_cols = encoded_data.select_dtypes(include = 'datetime64[ns]').columns.to_list()
date_cols
```

```python
for dc in date_cols:
    encoded_data[dc + '_Year'] = encoded_data[dc].dt.year
    encoded_data[dc + '_Month'] = encoded_data[dc].dt.month
    encoded_data[dc + '_Day'] = encoded_data[dc].dt.day

    encoded_data.drop(columns=[dc], axis=1, inplace=True)
```
